```python
# Fill numeric columns with mean
car.fillna(car.mean(numeric_only=True), inplace=True)

# Fill non-numeric columns with mode
for column in car.select_dtypes(include='object').columns:
    car[column].fillna(car[column].mode()[0], inplace=True)
```
